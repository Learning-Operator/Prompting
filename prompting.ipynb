```python
import cmbagent
from datasets import load_dataset
from dotenv import load_dotenv
import os
import pandas as pd
from pandas import DataFrame
```

```python
load_dotenv()
```

```python

def load_TP_public():
    ds = load_dataset("ZhiqiGao/TPBench")
    df = pd.DataFrame(ds["public"])
    return df


def prompt_wrapper(dataset: DataFrame, question_number) -> str:

    prompt = f"""

            The problem you are tasked with is as follows:

            {dataset.iloc[question_number]['problem']}

            Please follow these requirements when generating your code:

            {dataset.iloc[question_number]['code_answer_requirements']}

             """
    
    return prompt


def systematic(number_of_questions: int):

    df = load_TP_public()

    os.makedirs('./one_shot_solutions', exist_ok=True)

    for i in range(number_of_questions):
        
        os.makedirs(f'./one_shot_solutions/tpbench_oneshot_solution_q{i}', exist_ok=True)
        

        Cmbagent = cmbagent.one_shot(engineer_model="gpt-4o-mini",
                                    agent= 'engineer',
                                    work_dir=f'./one_shot_solutions/tpbench_oneshot_solution_q{i}',
                                    task = prompt_wrapper(df, i)
                                    )
        

systematic(2)

```
